```python
def jstacktrace(exc):
    if not hasattr(exc, 'classname') or exc.classname is None:
        return str(exc)
    return '' if not exc.stacktrace else '\n\tat '.join(exc.stacktrace)
```

The ImageJ1 window is a virtual window that doesn't immediately write back into the IJ2 structure upon changes.  There are no callbacks in Java to react to raw array data changes.

However, we found out that using a 3D image changing the hyperplane does have it write back to the IJ2 structure.  As such, once you change position (either programatically or in the window), it will correctly read the new data.  This does not work for 2D images.
