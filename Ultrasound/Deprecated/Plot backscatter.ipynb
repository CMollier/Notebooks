```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
csv_path = Path(str(Path.home()) +  '\\Box\\Research\\LINK\\System calibration\\2019-08-30 - BSC with 30 MHz and polynomial\\BSC for python.csv')
```

```python
df = pd.read_csv(csv_path)
```

```python
fig, ax = plt.subplots(figsize=[10, 7.5])
plt.loglog(df['Freq Faran'], df['BSC Faran'], color='black')
plt.loglog(df['Freq-30MHz'], df['BSC-30MHz'], color='green')
#plt.loglog(df['Freq-10Mhz'], df['BSC 10 MHz'], df['Freq-20MHz'], df['BSC 20 MHz'], df['Freq-30MHz'], df['BSC-30MHz'])
plt.grid(which='both')
plt.title('Hi-Frequency reference phantom')
plt.xlabel('Frequency (MHz)')
plt.ylabel('BSC ($cm^{-1} sr^{-1}$)', rotation=0, labelpad=20)
plt.xlim([10, 50])
plt.legend(['Faran theory', '30 MHz transducer'])
#plt.legend(['Faran theory', '10 MHz transducer and Saran phantom', '20 MHz transducer and Saran phantom', '30 MHz transducer and TPX phantom'])
save_path = csv_path = Path(str(Path.home()) +  '\\Box\\Research\\LINK\\System calibration\\2019-08-30 - BSC with 30 MHz and polynomial\\QUS-BSC_plot.png')
plt.savefig(str(save_path))
```
